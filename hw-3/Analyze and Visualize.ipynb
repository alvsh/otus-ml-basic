### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
b = np.arange(10, 50, 1)
b = np.flip(b)
print(b)
b_even = [x for x in b if x % 2==0]
print(b_even)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
N = np.random.randint(24, size=(6,4)) 
M = np.random.randint(12, size=(4,3)) 
print(N.shape)
print(M.shape)
K = N.dot(M)
print(K)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
d = np.random.randint(1, 100, size=(7,7))
print('Mean: ', np.mean(d))
print('STD: ', np.std(d))
normalizedData = d / np.linalg.norm(d)
print(normalizedData)
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
days = ('Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun')
for d in days:
    
    tipsd = tips[(tips['day'] == d)]
    if tipsd.empty == False:
        tipsdm = tipsd['total_bill'].mean()
    else:
        tipsdm = 0.0
    
    print(f' {d} : {tipsdm}')
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
bill_med = tips['total_bill'].median()
print(bill_med)
tips_filtered = tips[tips['total_bill'] > bill_med]
sex = ('Male', 'Female')


total_bill_male_mean = tips_filtered[tips_filtered['sex']=='Male']['total_bill'].mean()
total_bill_female_mean = tips_filtered[tips_filtered['sex']=='Female']['total_bill'].mean()

print(f' Total bill Male mean value = {total_bill_male_mean}')
print(f' Total bill Female mean value = {total_bill_female_mean}')
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker'] = tips['smoker'].apply(lambda x: 0 if x == 'No' else 1)
tips.tail()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
time_v = tips['time'].unique()
fig, axes = plt.pyplot.subplots(1, time_v.size)
for t_val, ax in zip(time_v, axes.flat):
    time_v_el = tips[tips['time'] == t_val]
    sns.histplot(time_v_el['tip'], kde=False, bins=30, ax=ax).set(title=t_val)
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
fig, axes = plt.pyplot.subplots(1, 2)
sns.scatterplot(data=tips[tips['sex']=='Male'], x="total_bill", y="tip", hue="smoker", ax=axes[0]).set(title='Male')
sns.scatterplot(data=tips[tips['sex']=='Female'], x="total_bill", y="tip", hue="smoker", ax=axes[1]).set(title='Female')
```
